```python
xymin, xymax = 0, 0

for rxn in tqdm(set_test):
    for molecule_type in ['R', 'T', 'P']:
        with np.load(f'Tests/outputs_dft/rxn{rxn}_{molecule_type}.npz') as res_dft:
            H_dft = res_dft['hessian']
            H_eigval_dft, H_eigvec_dft = np.linalg.eigh(H_dft)
            xymin = min(xymin, H_eigval_dft.min())
            xymax = max(xymax, H_eigval_dft.max())

xymin -= (xymax - xymin) * 0.05
xymax += (xymax - xymin) * 0.05
```

```python
colors = {'R': 'gold', 'T': 'brown', 'P': 'sandybrown'}
zorders = {'R': 2, 'T': 0, 'P': 1}
methods = ['nnani', 'nnt1x', 'nnt1xaug1m']
titles = ['Pre-trained (ANI-1)', 'Fine-tuned (T1x)', 'Fine-tuned (T1x-aug)']

fig, axs = plt.subplots(1, 3, figsize=(8.5, 3), sharex=True, sharey=True)

for i in range(3):
    res = {'R': [], 'T': [], 'P': []}
    for rxn in tqdm(set_test):
        for molecule_type in ['R', 'T', 'P']:
            with np.load(f'Tests/outputs_{methods[i]}/rxn{rxn}_{molecule_type}.npz') as res_nn:
                n_atom = res_nn['n_atom']
                H_nn = res_nn['hessian'].reshape(n_atom * 3, n_atom * 3) / (units.kcal/units.mol/units.Ang/units.Ang)
                H_eigval_nn, H_eigvec_nn = np.linalg.eigh(H_nn)
                H_eigval_left_nn, H_eigvec_left_nn = H_eigval_nn[0], H_eigvec_nn[:, 0]
            with np.load(f'Tests/outputs_dft/rxn{rxn}_{molecule_type}.npz') as res_dft:
                H_dft = res_dft['hessian']
                H_eigval_dft, H_eigvec_dft = np.linalg.eigh(H_dft)
            similarity_matrix = np.abs(cosine_similarity(H_eigvec_dft.T, H_eigvec_nn.T))
            row_ind, col_ind = linear_sum_assignment(-similarity_matrix)
            res[molecule_type].append((H_eigval_dft[row_ind], H_eigval_nn[col_ind]))
    res = {molecule_type: np.hstack(res[molecule_type]) for molecule_type in ['R', 'T', 'P']}
    for molecule_type in ['R', 'T', 'P']:
        axs[i].plot(res[molecule_type][0], res[molecule_type][1], '.', color=colors[molecule_type], zorder=zorders[molecule_type])
    res = {molecule_type: np.sqrt(np.mean(np.square(np.diff(res[molecule_type], axis=0)))) for molecule_type in ['R', 'T', 'P']}

    axs[i].axis('square')
    axs[i].plot([xymin, xymax], [xymin, xymax], ':', color='black')
    axs[i].set_title(titles[i], fontsize=10, pad=60)
    legend = axs[i].legend([f"R (RMSE = {res['R']:.0f})", f"TS (RMSE = {res['T']:.0f})", f"P (RMSE = {res['P']:.0f})"], loc='lower center', bbox_to_anchor=(0.5, 1.0))
    for legend_handle in legend.legend_handles:
        legend_handle.set_alpha(1)
axs[0].set_xlim(xymin, xymax)
axs[0].set_ylim(xymin, xymax)
axs[0].set_xticks([-2500, 0, 2500, 5000, 7500])
axs[0].set_yticks([-2500, 0, 2500, 5000, 7500])
axs[1].set_xlabel('DFT Hessian eigenvalue (kcal/mol/Å$^2$)')
axs[0].set_ylabel('NewtonNet Hessian\ncorresponding eigenvalue (kcal/mol/Å$^2$)')
plt.savefig('HessianEigenvalue_all.pdf', bbox_inches='tight')
plt.show()
```
