```python
model_paths = [
    ('Models/PretrainedModels/training_1/models/best_model_state.tar', 'Models/PretrainedModels/training_1/run_scripts/ani-newtonnet.yml'),
    ('Models/PretrainedModels/training_9/models/best_model_state.tar', 'Models/PretrainedModels/training_9/run_scripts/wb97_newton.yml'),
    ('Models/FinetunedModels/training_44/models/best_model_state.tar', 'Models/FinetunedModels/training_44/run_scripts/config050.yml'),
    ('Models/FinetunedModels/training_56/models/best_model_state.tar', 'Models/FinetunedModels/training_56/run_scripts/config0.yml'),
    ('Models/FinetunedModels/training_60/models/best_model_state.tar', 'Models/FinetunedModels/training_60/run_scripts/config0.yml'),
]
save_paths = [
    'Tests/outputs_nnani',
    'Tests/outputs_nnani1x',
    'Tests/outputs_nnt1x',
    'Tests/outputs_nnt1xaug1m',
    'Tests/outputs_nnt1xaug4m',
]
if not os.path.exists('Tests'):
    os.makedirs('Tests')

for model_path, save_path in zip(model_paths, save_paths):
    mlcalculator = MLAseCalculator(
        model_path=model_path[0],
        settings_path=model_path[1],
        method='autograd',
        disagreement='values',
        device='cpu',
        )

    if not os.path.exists(os.path.join('Tests', save_path)):
        os.makedirs(os.path.join('Tests', save_path))

    for rxn in tqdm(set_test):
        with np.load(f'Data/Transition1x/reaction_data/rxn{rxn}.npz') as data:
            irc = [0, -8, -7, -6, -5, -4, -3, -2, -1, 1]
            E_irc = data['E'][irc]
            R_irc = data['R'][irc]
            Z_irc = data['Z'][irc]
            molecule = Atoms(numbers=Z_irc[0], positions=R_irc[0])
            mlcalculator.calculate(molecule)
            np.savez_compressed(f'{save_path}/rxn{rxn}_R.npz', 
                n_atom=len(molecule),
                energy=mlcalculator.results['energy'],
                forces=mlcalculator.results['forces'],
                hessian=mlcalculator.results['hessian'],
                energy_disagreement=mlcalculator.results['energy_disagreement'],
                forces_disagreement=mlcalculator.results['forces_disagreement'],
                hessian_disagreement=mlcalculator.results['hessian_disagreement'])
            molecule = Atoms(numbers=Z_irc[-1], positions=R_irc[-1])
            mlcalculator.calculate(molecule)
            np.savez_compressed(f'{save_path}/rxn{rxn}_P.npz',
                n_atom=len(molecule),
                energy=mlcalculator.results['energy'],
                forces=mlcalculator.results['forces'],
                hessian=mlcalculator.results['hessian'],
                energy_disagreement=mlcalculator.results['energy_disagreement'],
                forces_disagreement=mlcalculator.results['forces_disagreement'],
                hessian_disagreement=mlcalculator.results['hessian_disagreement'])
            molecule = Atoms(numbers=Z_irc[np.argmax(E_irc)], positions=R_irc[np.argmax(E_irc)])
            mlcalculator.calculate(molecule)
            np.savez_compressed(f'{save_path}/rxn{rxn}_T.npz',
                n_atom=len(molecule),
                energy=mlcalculator.results['energy'],
                forces=mlcalculator.results['forces'],
                hessian=mlcalculator.results['hessian'],
                energy_disagreement=mlcalculator.results['energy_disagreement'],
                forces_disagreement=mlcalculator.results['forces_disagreement'],
                hessian_disagreement=mlcalculator.results['hessian_disagreement'])
            
print('done!')
```

```python
def write_in(Z, R, file):
    with open(file, 'w') as f:
        f.write('$molecule\n')
        f.write('0 1\n')
        for z, r in zip(Z, R):
            f.write(f'{z} {r[0]} {r[1]} {r[2]}\n')
        f.write('$end\n\n')
        f.write('$rem\n')
        f.write('job_type = freq\n')
        f.write('basis = 6-31G*\n')
        f.write('max_scf_cycles = 100\n')
        f.write('gen_scfman = true\n')
        f.write('xc_grid = 3\n')
        f.write('thresh = 14\n')
        f.write('s2thresh = 16\n')
        f.write('scf_algorithm = diis\n')
        f.write('resp_charges = true\n')
        f.write('symmetry = false\n')
        f.write('sym_ignore = true\n')
        f.write('method = wb97x\n')
        f.write('geom_opt_max_cycles = 200\n')
        f.write('$end\n')

def write_sh(file):
    with open(file, 'w') as f:
        f.write('#!/bin/bash -l\n')
        f.write('#SBATCH -n 2\n')
        f.write('#SBATCH -q regular\n')
        f.write('#SBATCH -t 4:00:00\n')
        f.write(f'#SBATCH -o outputs/{file}.out\n')
        f.write('#SBATCH -J T1x\n')
        f.write('#SBATCH -C cpu\n')
        f.write('#SBATCH -L SCRATCH\n')
        f.write('#SBATCH -A m2101\n\n')
        f.write('module load qchem/6.0.0\n')
        f.write('export QCTHREADS=40\n')
        f.write('export OMP_NUM_THREADS=40\n')
        f.write('export QCSCRATCH=/global/homes/e/ericyuan/20230310_Transition1x/20230606_QMHessians\n')
        f.write('export QCLOCALSCR=/tmp\n\n')
        f.write('rm -rf /tmp/scratch\n\n')
        f.write(f'time qchem -save inputs/{file}_R.in outputs/{file}_R.out outputs/{file}_R_scratch\n')
        f.write(f'cp outputs/{file}_R_scratch/HESS outputs/{file}_R.hess\n')
        f.write(f'rm -rf outputs/{file}_R_scratch\n\n')
        f.write(f'time qchem -save inputs/{file}_T.in outputs/{file}_T.out outputs/{file}_T_scratch\n')
        f.write(f'cp outputs/{file}_T_scratch/HESS outputs/{file}_T.hess\n')
        f.write(f'rm -rf outputs/{file}_T_scratch\n\n')
        f.write(f'time qchem -save inputs/{file}_P.in outputs/{file}_P.out outputs/{file}_P_scratch\n')
        f.write(f'cp outputs/{file}_P_scratch/HESS outputs/{file}_P.hess\n')
        f.write(f'rm -rf outputs/{file}_P_scratch\n\n')


if not os.path.exists('Tests/inputs_dft'):
    os.makedirs('Tests/inputs_dft')
if not os.path.exists('Tests/scripts_dft'):
    os.makedirs('Tests/scripts_dft')
for file in os.listdir('Data/Transition1x/reaction_data'):
    try:
        data = dict(np.load(f'Data/Transition1x/reaction_data/{file}'))
    except:
        continue
    irc = [0, -8, -7, -6, -5, -4, -3, -2, -1, 1]
    E_irc = data['E'][irc]
    R_irc = data['R'][irc]
    Z_irc = data['Z'][irc]
    write_in(Z_irc[0], R_irc[0], 'Tests/inputs_dft/'+file[:-4]+'_R.in')
    write_in(Z_irc[-1], R_irc[-1], 'Tests/inputs_dft/'+file[:-4]+'_P.in')
    write_in(Z_irc[np.argmax(E_irc)], R_irc[np.argmax(E_irc)], 'Tests/inputs_dft/'+file[:-4]+'_T.in')
    write_sh('Tests/scripts_dft/'+file[:-4]+'.sh')

print('done!')
```
