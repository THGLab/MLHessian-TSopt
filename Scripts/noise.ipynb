```python
def build_molecule(atom, bond):
    mol = RWMol()
    # atom = [a for a in atom if a != 'H']
    for i in range(len(atom)):
        mol.AddAtom(Chem.Atom(atom[i]))
    for i in range(len(atom)):
        for j in range(i + 1, len(atom)):
            if bond[i, j] == 1:
                mol.AddBond(i, j, Chem.BondType.SINGLE)
            elif bond[i, j] == 2:
                mol.AddBond(i, j, Chem.BondType.DOUBLE)
            elif bond[i, j] == 3:
                mol.AddBond(i, j, Chem.BondType.TRIPLE)
    Chem.rdmolops.SanitizeMol(mol)
    return mol

if not os.path.exists('Data/molecules_scratch'):
    os.mkdir('Data/molecules_scratch')
for rxn in tqdm(range(265)):
    if not os.path.exists(f'Data/molecules_scratch/{rxn:03}'):
        os.mkdir(f'Data/molecules_scratch/{rxn:03}')
    atom = Molecule.from_file(f'Data/sella_si/singlets/{rxn:03}.xyz').labels
    bond = np.array([line.split() for line in open(f'Data/molecules_kinbotprod_renamed/{rxn:03}_R.bond', 'r').readlines()], dtype=int)
    mol = build_molecule(atom, bond)
    assert mol.GetNumAtoms() == len(atom), f'Rxn{rxn:03}: expect {len(atom)} atoms, got {mol.GetNumAtoms()} atoms'
    AllChem.EmbedMolecule(mol)
    pos = mol.GetConformer().GetPositions()
    pos = ',\n'.join([f'\t"{atom[i]}", {pos[i][0]}, {pos[i][1]}, {pos[i][2]}' for i in range(len(atom))])
    tpl = template_from_name(names[rxn])
    msg = (
        f'{{\n'
        f'"title": "{rxn:03}",\n'
        f'"structure": [\n'
        f'{pos}\n'
        f'],\n'
        f'"charge": 0,\n'
        f'"mult": 1,\n'
        f'\n'
        f'"qc": "gauss",\n'
        f'"qc_command": "g16",\n'
        f'"families": "{tpl}",\n'
        f'\n'
        f'"queuing": "slurm",\n'
        f'"queue_name": "medium",\n'
        f'"queue_template": "qu.tmp",\n'
        f'"queue_job_limit": 100,\n'
        f'"ppn": 8,\n'
        f'"scratch": "/scratch/jzador",\n'
        f'"username": "jzador"\n'
        f'}}\n')
    with open(f'Data/molecules_scratch/{rxn:03}/{rxn:03}.json', 'w') as f:
        f.write(str(msg))
```
